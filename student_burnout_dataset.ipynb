```python
import pandas as pd
import random

data = []

for i in range(1, 301):
    age = random.randint(18, 23)
    year = random.randint(1, 4)

    study = random.randint(3, 12)
    sleep = random.randint(3, 9)
    screen = random.randint(2, 10)
    activity = random.randint(0, 5)

    subjects = random.randint(4, 7)
    assignments = random.randint(2, 7)
    pressure = random.randint(2, 10)
    social = random.randint(2, 9)

    # Stress logic (important)
    stress = (
    study * 0.3 +
    pressure * 0.3 +
    assignments * 0.15 +
    screen * 0.1 -
    sleep * 0.25 -
    activity * 0.1 -
    social * 0.1
)
    stress = max(1, min(10, stress))

    motivation = random.randint(3, 9)
    balance = random.randint(2, 9)

    # Burnout logic
    if stress >= 8:
        risk = "High"
        level = "High"
    elif stress >= 5:
        risk = "Medium"
        level = "Medium"
    else:
        risk = "Low"
        level = "Low"

    data.append([
        i, age, year, study, sleep, screen, activity,
        subjects, assignments, pressure, social,
        stress, motivation, balance, risk, level
    ])

columns = [
    "student_id","age","year_of_study","study_hours","sleep_hours",
    "screen_time","physical_activity","subjects_count",
    "assignments_per_week","academic_pressure","social_interaction",
    "stress_level","motivation_level","study_life_balance",
    "burnout_risk","burnout_level"
]

df = pd.DataFrame(data, columns=columns)

df.to_csv("student_burnout_dataset.csv", index=False)

print("Dataset generated successfully!")
```
